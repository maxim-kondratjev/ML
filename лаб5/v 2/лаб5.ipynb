```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from keras.layers import Dense
from keras.models import Sequential
import numpy as np
from scipy import misc
from PIL import Image

%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

#print(df) OPENCV-PYTHON IMPORT CV2 

```python
pixels = df.columns.drop('label')
y_column = 'label'
```

```python
df[pixels] = df[pixels].applymap(lambda x: 1 if x / 255>0 else 0)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 750):
    print(df.describe())
```

```python
train, validation = train_test_split(df, test_size=0.2)

x_train = train[pixels].values
y_train = train[y_column].values

x_val = validation[pixels].values
y_val = validation[y_column].values
print(x_train.shape, y_train.shape, x_val.shape, y_val.shape)
```

```python
y_train = y_train.reshape((y_train.shape[0], 1))
y_val = y_val.reshape((y_val.shape[0], 1))

print(x_train.shape, y_train.shape)
```

```python
#x_train
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
accuracy_score(y_val, lr.predict(x_val))
```
